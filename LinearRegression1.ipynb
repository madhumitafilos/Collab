```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the dataset

```python
url = 'https://raw.githubusercontent.com/madhumitafilos/Collab/main/auto-mpg.csv'
auto = pd.read_csv(url)
auto.head()
```

Show Scatter Plot

```python
plt.figure(figsize=(10,10))
plt.scatter(auto['weight'],auto['mpg'])
plt.plot(auto['weight'], (auto['weight']/ -105) + 55, c='red')
plt.title('Weight vs Mileage of Cars')
plt.xlabel('Weight')
plt.ylabel('Mileage')
plt.show()
```
