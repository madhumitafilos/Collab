To find the Car weight for Optimal mileage

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

Load dataset

```python
url = 'https://raw.githubusercontent.com/madhumitafilos/Collab/main/auto-mpg.csv'
auto = pd.read_csv(url)
auto.head()
```

Initialising 2 variables to represent feature(X) and target(y)

```python
X = auto[['weight']]
y = auto['mpg']
```

Create Linear Regression class and train the model

```python
LinearReg = LinearRegression()
LinearReg.fit(X,y)
```

Create a Scatter plot for Weight vs Mileage of cars

```python
plt.figure(figsize=(10,10))
plt.scatter(auto['weight'],auto['mpg'])
plt.plot(X, LinearReg.predict(X), c='red')
plt.title('Weight vs Mileage of cars')
plt.xlabel('Weight')
plt.ylabel('Mileage')
plt.show()
```
